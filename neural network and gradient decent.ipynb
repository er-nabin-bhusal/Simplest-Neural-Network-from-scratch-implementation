```python
import numpy as np 
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def sigmoid_prime(x):
    return sigmoid(x)*(1-sigmoid(x))
```

```python
def neural_compute(m1,m2,w1,w2,b):
    return m1*w1+m2*w2+b
```

```python
data = [[1,1,1],[1,0,0],[0,1,0],[0,0,0]]
```

```python
w1 = np.random.randn()
w2 = np.random.randn()
b = np.random.randn()
learning_rate=0.01
```

```python
costs = []
for i in range(39000):
    ri = np.random.randint(len(data))
    point = data[ri]
    summation = neural_compute(point[0],point[1],w1,w2,b)
    pred = sigmoid(summation)
    target = point[2]
    cost = np.square(pred-target)
    dcost_pred = 2 * (pred-target)
    dpred_dz = sigmoid_prime(summation)
    dz_dw1 = point[0]
    dz_dw2 = point[1]
    dz_db = 1
    dcost_dz = dcost_pred * dpred_dz
    
    dcost_dw1 = dcost_dz * dz_dw1
    dcost_dw2 = dcost_dz * dz_dw2
    dcost_db = dcost_dz * dz_db
    w1 = w1 - learning_rate * dcost_dw1
    w2 = w2 - learning_rate * dcost_dw2
    b = b - learning_rate * dcost_db
    
    if i % 100 == 0:
        cost_sum = 0 
        for j in range(len(data)):
            p = data[j]
            z = p[0]*w1+p[1]*w2+b
            pred= sigmoid(z)
            target = p[2]
            cost_sum += np.square(pred-target)
        print(cost_sum)
        costs.append(cost_sum)
```
